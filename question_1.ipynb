```python
def search(txt):
    msg = []
    for a in txt.strip().split('.'):
        if word in a:
            msg.append(a)
    
    if len(msg) > 0:
        return msg
    else:
        return 'Word not found'
```

```python
def add_news(x, table):
    news_url = x['href'].split('/')
    news_url[0] = 'https://news.google.com'
    news_url = '/'.join(news_url)

    news = Article(news_url, language = 'en')

    news.download()
    news.parse()
    news.nlp()

    msg = search(news.text)
    
    table = table.append({'title' : news.title, 'summary' : news.summary, 'date-time' : news.publish_date, 'url' : news_url}, ignore_index = True)
    return table, msg
```
